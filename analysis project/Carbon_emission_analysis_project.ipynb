**Does the vehicle class have impact on co2 emission**
- NULL hypothesis : there  is no difference in co2 emission across different vehicle class.

```python
import scipy.stats as stats
groups = [data[data['VehicleClass'] == vehicle_class]['CO2Emissions'] for vehicle_class in data['VehicleClass'].unique()]#selecting and grouping all the vehicle classes.
f_stat, p_value = stats.f_oneway(*groups)#calculating oneway anova
print("F-statistic:",f_stat, "p-value: ",p_value)#printing the statitic value and p value
```

**By performing one way anova to find out how vehicle class have variations  in co2 emission.**
- The F-statistic measures how much the means of CO2 emissions differ between the vehicle classes  relative to the variability within each group.
- An F-statistic of 266.02 is quite large, it shows that there is a significant difference in CO2 emissions across the different vehicle classes.
- Since this p-value is much smaller than 0.05, we reject the null hypothesis and conclude that there is a significant difference in CO2 emissions between the  different vehicle classes.

**Does the transmission type have impact on co2 emission**
-  NULL hypothesis : there  is no difference in co2 emission across different transmission types.

```python
import scipy.stats as stats
groups = [data[data['Transmission'] == vehicle_class]['CO2Emissions'] for vehicle_class in data['Transmission'].unique()]
f_stat, p_value = stats.f_oneway(*groups)
print(f"F-statistic: {f_stat}, p-value: {p_value}")
```

- The F-statistic measures how much the means of CO2 emissions differ between the transmission typesrelative to the variability within each group.
- An F-statistic of 103.703 is quite large, it shows that there is a significant difference in CO2 emissions across the different transmission types.
- Since this p-value is much smaller than 0.05, we reject the null hypothesis and conclude that there is a significant difference in CO2 emissions between the  different transmission types.
